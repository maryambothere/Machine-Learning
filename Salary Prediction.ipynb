<h1>Simple Linear Regression: Years of Experience vs Salary</h1>

<h2>Objective</h2>
<p>To predict an employee’s salary based on their years of work experience using a simple linear regression model.</p>

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
#importing dataset
df = pd.read_csv('Salary_dataset.csv')
```

```python
#checking dataset
df.head()
```

```python
#dropping unnecessary column
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df.head()
```

```python
#Visualizing
sns.scatterplot(
    data = df,
    x = 'YearsExperience',
    y = 'Salary'
)
```

```python
#Setting Variables
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
#Training the model
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Evaluating the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse) 
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```
